In questo student tutorial vedremo come poter utilizzare le api ufficiali di Tumblr per Python.

## Come iniziare
Per poter procedere ti serve:
1. Un account di Tumblr
- Un'app registrata per ottenere i codici OAuth
- Installare il modulo python per usare le API di Tumblr

**1. REGISTRARSI**

Per registrarti ti basta andare al seguente indirizzo:
- https://www.tumblr.com/register

**2. Creare un'app e ottenere i codici OAuth**

Una volta registrato potrai creare una nuova app al seguente indirizzo:
- https://www.tumblr.com/oauth/apps

Per ottenere i codici OAuth dovrai:

1. Cliccare sul tasto **Registra l'applicazione**



2. Specificare i dati dell'applicazione



Non è richiesto indicare molti dati. Se non dovessi avere degli URL adatti, indica l'URL del tuo profilo come **placeholder**: potrai sempre modificare queste impostazioni in un secondo momento. Nella descrizione basta indicare brevemente che lo si userà per scopi accademici/studio personale.

**N.B.** Per nuovi account sono imposti dei limiti di utilizzo.



La nostra app non richiede di dover effettuare una richiesta di rimozione dei limiti. In ogni caso ciò sarà sempre possibile farlo e lo potrai anche richiedere da subito, non appena creata l'app, come indica l'immagine sottostante.



Altre limitazioni possono essere attenzionate al link:

- https://www.tumblr.com/docs/en/api/v2#rate-limits

3. Accedere alla pagina di gestione dell'app

Potrai farlo andando allo stesso url delle app indicato prima:

- https://www.tumblr.com/oauth/apps

Dopo aver cliccato sull'icona dell'app che vuoi modificare, potrai trovare i tuoi codici OAuth sul pannello laterale destro.



**3. Installare il modulo Python per accedere alle API di Tumblr**

Per installare il modulo dovremo usare uno dei seguenti comandi:

1. conda install -c akode pytumblr
- pip install pytumblr

**N.B.** Ho personalmente avuto problemi ad installare il pacchetto usando il comando di conda e quindi ho provveduto ad installarlo manualmente scaricando il pacchetto dal sito ufficiale (https://anaconda.org/akode/pytumblr/files).
